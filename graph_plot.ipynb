```python
import pickle as pkl
import matplotlib.pyplot as plt
import os
```

```python


plt.figure(figsize=(12, 12))

plt.rc('axes', titlesize=18) # fontsize of the axes title 
plt.rc('axes', labelsize=18) # fontsize of the x and y labels 
plt.rc('xtick', labelsize=15) # fontsize of the tick labels 
plt.rc('ytick', labelsize=15) # fontsize of the tick labels 
plt.rc('legend', fontsize=14) # legend fontsize 

plt.title('Last updated accuracy by prefix length')
plt.ylabel('Accuracy')
plt.xlabel('Prefix length')
plt.xticks(rotation='45')
axes = plt.gca()
axes.set_ylim([0.4, 1.05])

with open('./result/off_dt_acc.pkl' ,'rb') as f:
    data = pkl.load(f)
plt.plot(data[0],data[1],'-*',color='green',linewidth=4, linestyle='--',label = 'OFF DT',markersize=10)

with open('./result/off_rf_acc.pkl' ,'rb') as f:
    data = pkl.load(f)
plt.plot(data[0],data[1],'-*',color='purple',linewidth=4, linestyle='--',label = 'OFF RF',markersize=10)


with open('./result/efdt_lastacc.pkl' ,'rb') as f:
    data = pkl.load(f)
plt.plot(data[0],data[1],'-o',color='red',linewidth=4, label = 'EFDT',markersize=10)

with open('./result/hatc_lastacc.pkl' ,'rb') as f:
    data = pkl.load(f)
plt.plot(data[0],data[1],'-o',color='blue',linewidth=4, label = 'HATC',markersize=10)
with open('./result/htc_lastacc.pkl' ,'rb') as f:
    data = pkl.load(f)
plt.plot(data[0],data[1],'-o',color='orange',linewidth=4, label = 'HTC',markersize=10)

plt.legend()
plt.tight_layout()
# plt.savefig('./img/online last_acc_streaming.png')
plt.savefig('./img/last_acc_streaming.png')
```
